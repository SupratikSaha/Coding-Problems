    Problem Statement

    We are given a binary tree (with root node root), a target node, and an integer value K.

    Return a list of the values of all nodes that have a distance K from the target node.  The answer can be returned in any order.



    Example 1:

    Input: root = [3,5,1,6,2,0,8,null,null,7,4], target = 5, K = 2

    Output: [7,4,1]

    Explanation: 
    The nodes that are a distance 2 from the target node (with value 5)
    have values 7, 4, and 1.



    Note that the inputs "root" and "target" are actually TreeNodes.
    The descriptions of the inputs above are just serializations of these objects.



    Note:

        The given tree is non-empty.
        Each node in the tree has unique values 0 <= node.val <= 500.
        The target node is a node in the tree.
        0 <= K <= 1000.

# DFS - O(N) runtime, O(N) space

```python
from typing import List

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def distanceK(self, root: TreeNode, target: TreeNode, K: int) -> List[int]:
        res = []
        self.distanceKRecursive(root, target, K, res, float('inf'))
        return res
        
        
    def distanceKRecursive(self, root: TreeNode, target: TreeNode, K: int, res: List[int], currDistance: int) -> None:
        if not root: return float('inf')
        if K == currDistance: res.append(root.val)
        
        if root.val == target.val:
            currDistance = 0
            if K == 0:
                res.append(root.val)
                return float('inf')
            
        left = 1 + self.distanceKRecursive(root.left, target, K, res, 1 + currDistance)
        if left != float('inf'):
            right = 1 + self.distanceKRecursive(root.right, target, K, res, 1 + left)
        else:
            right = 1 + self.distanceKRecursive(root.right, target, K, res, 1 + currDistance)
            
        if right != float('inf'):
            left = 1 + self.distanceKRecursive(root.left, target, K, res, 1 + right)
        
        if left == K or right == K:
            res.append(root.val)
        
        if root.val == target.val: return 0
        else: return min(left, right)
```

# DFS - Percolate Distance - O(N) runtime, O(N) space

```python
from typing import List

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def distanceK(self, root, target, K):
        ans = []

        # Return distance from node to target if exists, else -1
        # Vertex distance: the # of vertices on the path from node to target
        def dfs(node):
            if not node:
                return -1
            elif node is target:
                subtree_add(node, 0)
                return 1
            else:
                L, R = dfs(node.left), dfs(node.right)
                if L != -1:
                    if L == K: ans.append(node.val)
                    subtree_add(node.right, L + 1)
                    return L + 1
                elif R != -1:
                    if R == K: ans.append(node.val)
                    subtree_add(node.left, R + 1)
                    return R + 1
                else:
                    return -1

        # Add all nodes 'K - dist' from the node to answer.
        def subtree_add(node, dist):
            if not node:
                return
            elif dist == K:
                ans.append(node.val)
            else:
                subtree_add(node.left, dist + 1)
                subtree_add(node.right, dist + 1)

        dfs(root)
        return ans
```
